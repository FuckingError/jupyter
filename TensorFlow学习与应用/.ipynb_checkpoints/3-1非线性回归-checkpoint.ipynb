```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#随机生成 -0.5到0.5 200个数 化成多行一列
x_data = np.linspace(-0.5,0.5,200)[:,np.newaxis]
#干扰
noise = np.random.normal(0,0.02,x_data.shape)
#生成 非线性方程
y_data = np.square(x_data)+noise

#两个占位符  一列
x = tf.placeholder(tf.float32,[None,1])
y = tf.placeholder(tf.float32,[None,1])

#中间层
#权重参数
# ？ 权重参数如何初始化？  使用tf.Variable  10个神经元 
Weight_L1 = tf.Variable(tf.random_normal([1,10]))
#偏置
biases_L1 = tf.Variable(tf.zeros([1,10]))
#得出结果
Weight_Plus_b_L1 = tf.matmul(x,Weight_L1)+biases_L1
# !别忘记激活函数
L1 = tf.nn.tanh(Weight_Plus_b_L1)

#输出层  10->1
Weight_L2 = tf.Variable(tf.random_normal([10,1]))
biases_L2 = tf.Variable(tf.zeros([1,1]))  # !要和矩阵乘积的结果  数目一致
#得出结果
Weight_Plus_b_L2 = tf.matmul(L1,Weight_L2)+biases_L2
#激活函数
prediction = tf.nn.tanh(Weight_Plus_b_L2)

#使用二次代价函数 计算loss
loss = tf.reduce_mean(tf.square(prediction-y))
#梯度下降
train_step = tf.train.GradientDescentOptimizer(0.1).minimize(loss)

#session
with tf.Session() as sess:
    #参数初始化
    sess.run(tf.global_variables_initializer())
    for _ in range(20000):
        sess.run(train_step,feed_dict={x:x_data,y:y_data})  #！ 运行这一段graph 需要x 和 y
    #获得预测值   
    prediction_value = sess.run(prediction,feed_dict={x:x_data})  #只需要x
    plt.figure()
    plt.scatter(x_data,y_data) #!散点
    plt.plot(x_data,prediction_value,'r*',lw=5) #plot 常见属性 lw宽度
    plt.show()
```
