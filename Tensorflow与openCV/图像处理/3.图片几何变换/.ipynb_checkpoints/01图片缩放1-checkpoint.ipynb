```python
#步骤 1.load 2.info 3.resize 4.check
import cv2
```

```python
for i in range(1,305):
    img = cv2.imread('pos/ ('+str(i)+').jpg',1) #../上层目录  ./当前目录
    imgInfo = img.shape  #获取图片 长 宽 通道数
    height = imgInfo[0]
    width = imgInfo[1]
    print(height,width)
    mode = imgInfo[2]
    dstHeight = int(height*1)
    dstWidth = int(width*1)
    #调用resize方法  有四种: 最近邻域插值 双线性插值 像素关系重采样 立方插值\
    #一般选择第二种
    dst = cv2.resize(img,(dstWidth,dstHeight))
    cv2.imwrite('n/'+str(i)+'.jpg',dst)
print('end')

# 1.放大 缩小两种   2.等比例缩放 非等比例缩放
```

### 等比例缩放

```python
dstHeight = int(height*0.5)
dstWidth = int(width*0.5)
#调用resize方法  有四种: 最近邻域插值 双线性插值 像素关系重采样 立方插值\
#一般选择第二种
dst = cv2.resize(img,(dstWidth,dstHeight))
cv2.imshow('image',dst)
cv2.waitKey()
```
