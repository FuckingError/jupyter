#### 边缘检测涉及到的几个概念：
    sobel算法的几个原理
    1.算子模板:有水平方向 和 竖直方向两种
        竖直:         水平：        
        [1 2 1       [1 0 -1
         0 0 0        2 0 -2
        -1 -2 -1]     1 0 -1]
        
    2.图片卷积
        举例: [1 2 3 4][a b c d]  = [1*a+2*b+3*c+4*d] =dst 可得到一个计算结果 这个就是梯度
    3.阈值判决
        梯度分水平方向 和 竖直方向两种情况
        使用根号下 a²+b² 计算梯度的方向
        sqrt(a*a+b*b) = f   a是水平方向的计算结果 b是竖直方向上的计算结果
        开方运算后得到的结果就是阈值 
        将阈值与门限th进行比较
        如果 f>th 则是边缘 f<th就是非边缘

```python
import cv2
import numpy as np
import random
import math
img = cv2.imread('test.jpg',1)
imgInfo = img.shape
height = imgInfo[0]
width = imgInfo[1]
```

```python
#1.灰度处理
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
dst = np.zeros((height,width,1),np.uint8)


#2.定义两个算子 水平和竖直方向  视频中的计算太烦了...  我直接定义两个算子 然后调用numpy矩阵元素相乘 然后平直操作 求和
oy = np.mat([[1,2,1],[0,0,0],[-1,-2,-1]]) #operator x
ox = np.mat([[1,0,-1],[2,0,-2],[1,0,-1]])

#注意遍历矩阵时会发生溢出问题
#这里的卷积的算子是3*3的矩阵 所以遍历时要减去2
for i in range(0,height-2):
    for j in range(0,width-2):
        if j%100==0:
            print(i," ",j)
        #3.计算梯度
        dsty = np.sum(np.ravel(np.multiply(gray[i:i+3,j:j+3],oy)))
        dstx = np.sum(np.ravel(np.multiply(gray[i:i+3,j:j+3],ox)))
        #4.计算阈值
        grad = math.sqrt(dsty*dsty+dstx*dstx)
        #5.阈值判决
        if grad>50:
            dst[i,j] = 255
        else:
            dst[i,j] = 0
cv2.imwrite('zmy__.jpg',dst)        
        
        
```
