    1.1  pos文件夹正样本 包含所预测目标   neg文件夹中不包含obj
    1.2获取样本的途径
    1.3正样本中尽可能的多样 环境干扰
    正样本和负样本的比例为 1:2 或 1：3之间


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#步骤：1.参数设置 2.创建hog 3.创建svm 4.计算hog 5.label 6.train 7.pred 8.draw

#1.参数定义
PosNum = 304
NegNum = 437
winSize = (64,128)
blockSize = (16,16) #105个
blockStride= (8,8) #步长
cellSize = (8,8) #4个cell  cell不可移动
nbin = 9 #9个bin  一个3780个
```

```python
#2.创建hog
hog = cv2.HOGDescriptor(winSize,blockSize,blockStride,cellSize,nbin)
#3.创建svm
svm = cv2.ml.SVM_create()
#4.计算 hog
    #4.1 定义特征维度
featureNum = int(((128-16)/8+1)*((64-16)/8+1)*4*9)#3780
    #4.2 定义特征数组
featureArray = np.zeros(((PosNum+NegNum),featureNum),np.float32) #行数 是样本总数 列数是特征维度
    #4.3 标签数组
labelArray = np.zeros(((PosNum+NegNum),1),np.int32)
```

```python
#svm在进行学习的时候 计算出的hog才是真正的样本
#5.计算正样本Hog 和 给标签
for i in range(0,PosNum):
    fileName = 'pos/'+str(i+1)+'.jpg'
    img = cv2.imread(fileName)
    print(img.shape)
    '''
    #5.1计算
    hist = hog.compute(img,(8,8))#hot值
    #5.2将计算出来的Hog放入特征向量中

    for j in range(0,featureNum):
        featureArray[i,j] = hist[j]
    #5.3 给标签
    labelArray[i,0] = 1 #这里全部都是正样本  1
'''
```

```python
# 计算负样本Hog 给标签
for i in range(0,NegNum):
    fileName = 'neg/'+str(i+1)+'.jpg'
    img = cv2.imread(fileName)
    #5.1计算
    hist = hog.compute(img,(8,8))#hot值
    #5.2将计算出来的Hog放入特征向量中
    for j in range(j,featureNum):
        featureArray[PosNum+i,j] = hist[j]
    #5.3 给标签
    labelArray[i+PosNum,0] = -1 #这里全部都是负样本  -1
```

```python
    
#SVM参数设置
svm.setType(cv2.ml.SVM_C_SVC)
svm.setKernel(cv2.ml.SVM_LINEAR) #线性内核
svm.setC(0.01)

#6.训练
ret = svm.train(featureArray,cv2.ml.ROW_SAMPLE,labelArray)

#7.预测
alpha = np.zeros((1),np.float32)
rho = svm.getDecisionFunction(0,alpha)
print(rho)
print(alpha)

alphaArray = np.zeros((1,1),np.float32)
supportVArray = np.zeros((1,featureNum),np.float32) #支持向量机数组
resultArray = np.zeros((1,featureNum),np.float32)

alphaArray[0,0] = alpha
#计算resultArray
resultArray = -1*alphaArray*supportVArray

#detect
myDetect = np.zeros((3781),np.float32)
for i in range(0,3780):
    myDetect[i] = resultArray[0,i]
myDetect[3780] = rho[0]
#构建myhog
myHog = cv2.HOGDescriptor()
myHog.setSVMDetcetor(myDetect)
#预测
imageSrc = cv2.imread('115.jpg',1)

objs = myHog.detectMultiScale(imageSrc,0,(8*8),(32*32),1.05,2) #5. 缩放比例 
#获取检测出来的目标
x = int(objs[0][0][0])
y = int(objs[0][0][1])
w = int(objs[0][0][2])
h = int(objs[0][0][3])
cv2.rectangle(imageSrc,(x,y),(x+w,y+h),(255,0,0),2)#3.颜色 4.宽度
cv2.imshow('dst',imageSrc)
cv2.waitKey(0)
```
